```python
from pyspark import SparkContext

logFile = "test.txt"  # Should be some file on your system
sc = SparkContext("local", "Simple App")
logData = sc.textFile(logFile).cache()

output = logData.flatMap(lambda s: s.split("|")) \
    .map(lambda s: s.split(",")) \
    .map(lambda s: (s[0], int(s[1]))) \
    .filter(lambda s: s[1] > 0) \
    .reduceByKey(lambda x, y: x + y).sortBy(lambda s: s[1],ascending=False).collect()

for i in output:
    print(i)
```

```python
from pyspark import SparkContext

logFile = "test.txt"  # Should be some file on your system
sc = SparkContext("local", "Simple App")
logData = sc.textFile(logFile).cache()

allsum = logData.flatMap(lambda s: s.split("|")) \
    .map(lambda s: s.split(",")) \
    .map(lambda s: (s[0], float(s[1]))) \
    .filter(lambda s: s[1] > 0) \
    .reduceByKey(lambda x, y: x + y)

count = logData.flatMap(lambda s: s.split("|")) \
    .map(lambda s: s.split(",")) \
    .map(lambda s: (s[0],float(s[1]))) \
    .filter(lambda s: s[1] > 0) \
    .map(lambda s: (s[0], 1))\
    .reduceByKey(lambda x, y: x + y)

ans = allsum.join(count).mapValues(lambda value: value[0]/value[1]).sortBy(lambda s: s[1]).collect()
for key, value in ans:
    print(key, value)

```
